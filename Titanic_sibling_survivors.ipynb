We now set parameters for getting a bar graph of the attributes with respect to the suvivors.

```python
def barChart(feature):
    survived = train[train['Survived']==1][feature].value_counts()
    dead = train[train['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

Now, we see below the bar graph between the SibSp i.e. the SiblingSpouse number and the surviors and non survivors for each.

We will now combine the train and the test dataset and separate the title from the Name of each Passenger.

```python
trainTestData = [train, test]
for data in trainTestData:
    data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
train['Title'].value_counts()
```

```python
test['Title'].value_counts()
```

We now map the title to numerical value so as to get rid of a string feature from the dataframe.
We convert the Mr to 0, Miss to 1, Mrs to 2 and Master, Dr, Rev, Col, Major, Mlle, Countess, Ms, Lady, Jonkheer, Don, Dona, Mme, Capt, Sir to 3

```python
titleMap = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
for data in trainTestData:
    data['Title'] = data['Title'].map(titleMap)
```

```python
train.head()
```

```python
test.head()
```

```python
train.drop('Name', axis=1, inplace=True)
test.drop('Name', axis=1, inplace=True)
```

```python
train.head()
```

```python
test.head()
```

Now we convert the sex feature to numerical value -- Male to 0 and Female to 1

```python
sexMap = {"male": 0, "female": 1}
for data in trainTestData:
    data['Sex'] = data['Sex'].map(sexMap)
```

```python
train.head()
```

```python
test.head()
```

As the Age column has Null values we can fill them by taking an average of the people in the same title group

```python
train["Age"].fillna(train.groupby("Title")["Age"].transform("median"), inplace=True)
test["Age"].fillna(test.groupby("Title")["Age"].transform("median"), inplace=True)
```

```python
train.head(30)
train.groupby("Title")["Age"].transform("median")
```

```python
train.info()
```

```python
test.info()
```

Now we will group the Age column by age groups and map them as follows -
upto 17 --> 0
17 to 25 --> 1
26 to 40 --> 2
40 to 60 --> 3
above 60 --> 4

```python
for data in trainTestData:
    data.loc[ data['Age'] <= 16, 'Age'] = 0,
    data.loc[(data['Age'] > 16) & (data['Age'] <= 26), 'Age'] = 1,
    data.loc[(data['Age'] > 26) & (data['Age'] <= 36), 'Age'] = 2,
    data.loc[(data['Age'] > 36) & (data['Age'] <= 62), 'Age'] = 3,
    data.loc[ data['Age'] > 62, 'Age'] = 4
```

```python
train.head()
```

```python
test.head()
```

We will now replace the null values in Embarked column to Southampton i.e. S as we will not need this majorly and S is present in the highest ratio. Also we map the column as follows:
S --> 0
C --> 1
Q --> 2

```python
for data in trainTestData:
    data['Embarked'] = data['Embarked'].fillna('S')
```

```python
train.head()
```

```python
embarkedMap = {"S": 0, "C": 1, "Q": 2}
for data in trainTestData:
    data['Embarked'] = data['Embarked'].map(embarkedMap)
```

For the missing values in Fare we will be filling it using the median of fares for each PClass.

```python
train["Fare"].fillna(train.groupby("Pclass")["Fare"].transform("median"), inplace=True)
test["Fare"].fillna(test.groupby("Pclass")["Fare"].transform("median"), inplace=True)
train.head()
```

```python
test.head(30)
```

We will now map fares as per the following ranges:
upto 18 --> 0
18 to 30 --> 1
31 to 100 --> 2
above 100 --> 3

```python
for data in trainTestData:
    data.loc[ data['Fare'] <= 17, 'Fare'] = 0,
    data.loc[(data['Fare'] > 17) & (data['Fare'] <= 30), 'Fare'] = 1,
    data.loc[(data['Fare'] > 30) & (data['Fare'] <= 100), 'Fare'] = 2,
    data.loc[ data['Fare'] > 100, 'Fare'] 
```

```python
train.head()
```

```python
test.head()
```

Now for the data in Cabin we will convert it to only one letter and then use the first letters to group them

```python
for data in trainTestData:
    data['Cabin'] = data['Cabin'].str[:1]
```

```python
cabinMap = {"A": 0, "B": 0.4, "C": 0.8, "D": 1.2, "E": 1.6, "F": 2, "G": 2.4, "T": 2.8}
for data in trainTestData:
    data['Cabin'] = data['Cabin'].map(cabinMap)
```

```python
train.head()
```

```python
test.head()
```

We will fill in the missing cabin information using the PClass column and by taking the median of the group

```python
train["Cabin"].fillna(train.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
test["Cabin"].fillna(test.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
```

```python
train.head()
```

Now we can move on to the family features i.e. Parch and the SibSp relations, We will calculate them as one variable for family and one for Siblings. 

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
test["FamilySize"] = test["SibSp"] + test["Parch"] + 1
```

```python
train["sibling"] = train["SibSp"]
test["sibling"] = test["SibSp"]
```

Also we can view the graph for relation between siblings and survivors in a different way

```python
fac = sns.FacetGrid(train, hue="Survived",aspect=4)
fac.map(sns.kdeplot,'sibling',shade= True)
fac.set(xlim=(0, train['sibling'].max()))
fac.add_legend()
plt.xlim(0)
```

We can now Map the family size using a scale between 0 and 4 and same for siblings

```python
familyMap = {1: 0, 2: 0.4, 3: 0.8, 4: 1.2, 5: 1.6, 6: 2, 7: 2.4, 8: 2.8, 9: 3.2, 10: 3.6, 11: 4}
for data in trainTestData:
    data['FamilySize'] = data['FamilySize'].map(familyMap)
```

```python
siblingMap = {0: 0, 1: 0.5, 2: 1, 3: 1.5, 4: 2, 5: 2.5, 6: 3, 7: 3.5, 8: 4}
for data in trainTestData:
    data['sibling'] = data['sibling'].map(siblingMap)
```

```python
train.head()
```

Now we will drop features we will not be needing

```python
features = ['Ticket', 'SibSp', 'Parch']
train = train.drop(features, axis=1)
test = test.drop(features, axis=1)
train = train.drop(['PassengerId'], axis=1)
```

```python
trainData = train.drop('Survived', axis=1)
target = train['Survived']

trainData.shape, target.shape
```

```python
trainData.head(15)
```

## Modelling of the data and calculating accuracy by  applying classifiers

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
train.info()
```

### K-fold Cross validation and kNN algorithm

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
clf = KNeighborsClassifier(n_neighbors = 14)
scoring = 'accuracy'
score = cross_val_score(clf, trainData, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# kNN Score
round(np.mean(score)*100, 2)
```

### Descision Tree Classifier

```python
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, trainData, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# decision tree Score
round(np.mean(score)*100, 2)
```

### Random Forest Classifier

```python
clf = RandomForestClassifier(n_estimators=14)
scoring = 'accuracy'
score = cross_val_score(clf, trainData, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Random Forest Score
round(np.mean(score)*100, 2)
```

### Naive Bayes Classifier

```python
clf = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clf, trainData, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Naive Bayes Score
round(np.mean(score)*100, 2)
```

### SVM (Support Vector Machine) Classifier

```python
clf = SVC()
scoring = 'accuracy'
score = cross_val_score(clf, trainData, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

##### Since SVM has the highest mean accuracy we will use that to predict the survivors on the test data

```python
clf = SVC()
clf.fit(trainData, target)

testData = test.drop("PassengerId", axis=1).copy()
prediction = clf.predict(testData)
```

```python
finalPrediction = pd.DataFrame({
        "PassengerId": test["PassengerId"],
        "Sibling": test["sibling"],
        "Survived": prediction
    })

finalPrediction.to_csv('FinalPrediction.csv', index=False)
```

```python
finalPrediction = pd.read_csv('FinalPrediction.csv')
finalPrediction.head()
```

Now we will show the relation between the number of siblings and the survivors on the test data

```python
fac = sns.FacetGrid(finalPrediction, hue="Survived",aspect=4)
fac.map(sns.kdeplot,'Sibling',shade= True)
fac.set(xlim=(0, finalPrediction['Sibling'].max()))
fac.add_legend()
plt.show()
```
